```python
%%bigquery
LOAD DATA INTO `qwiklabs-gcp-01-4b0fd1e23a0f.fraud_data.fraud_data_raw`
FROM FILES (
  -- Specify the file format
  format = 'CSV',
  -- Provide the full GCS path to the file
  uris = ['gs://labs.roitraining.com/data-to-ai-workshop/fraud_data_raw.csv']
)
```

```python
%%bigquery
CREATE OR REPLACE TABLE `qwiklabs-gcp-01-4b0fd1e23a0f.fraud_data.fraud_training_data` AS (SELECT
    CASE WHEN Employment_Status = 'Employed' THEN 1 ELSE 0 END AS is_employed,
    CASE WHEN Employment_Status = 'Unemployed' THEN 1 ELSE 0 END AS is_unemployed,
    CASE WHEN Employment_Status = 'Self-Employed' THEN 1 ELSE 0 END AS is_self_employed,

    CASE WHEN Device_Type = 'Mobile' THEN 1 ELSE 0 END AS is_device_mobile,
    CASE WHEN Device_Type = 'Desktop' THEN 1 ELSE 0 END AS is_device_desktop,
    CASE WHEN Device_Type = 'Tablet' THEN 1 ELSE 0 END AS is_device_tablet,

    CASE
      WHEN Age BETWEEN 18 AND 24 THEN 1
      ELSE 0
    END AS age_bin_18_24,
    CASE
      WHEN Age BETWEEN 25 AND 34 THEN 1
      ELSE 0
    END AS age_bin_25_34,
    CASE
      WHEN Age BETWEEN 35 AND 44 THEN 1
      ELSE 0
    END AS age_bin_35_44,
    CASE
      WHEN Age BETWEEN 45 AND 54 THEN 1
      ELSE 0
    END AS age_bin_45_54,
    CASE
      WHEN Age >= 55 THEN 1
      ELSE 0
    END AS age_bin_55_plus,

    SAFE_DIVIDE(Income, Amount_Requested) AS income_to_amount_requested_ratio,

    COALESCE(
      DATE_DIFF(
        CURRENT_DATE(),
        Previous_Assistance_Date,
        DAY
      ),
      0
    ) AS time_since_previous_assistance_days,

    CAST(Supporting_Doc_Verified AS INT64) AS supporting_doc_verified_int,
    CAST(Previous_Assistance_Received AS INT64) AS previous_assistance_received_int

    FROM `qwiklabs-gcp-01-4b0fd1e23a0f.fraud_data.fraud_data_raw`)
```
